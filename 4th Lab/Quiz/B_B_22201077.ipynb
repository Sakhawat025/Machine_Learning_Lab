```python

# Task 1: Create a 1D array of 10 random integers between 1 and 50
import numpy as np
arr = np.random.randint(1, 51, size=10)
arr
```

```python
# Task 2: Reshape the above 10-element array into a 2x5 2D matrix
arr_reshaped = arr.reshape(2, 5)
arr_reshaped
```

```python
# Task 3: Create a DataFrame from a dictionary containing names and marks of 3 students
import pandas as pd
data = {'Name': ['Sakhawat', 'Rakib', 'Shahin'], 'Marks': [85, 90, 88]}
df = pd.DataFrame(data)
df
```

```python
# Task 4: Use the `head()` function to display only the first 2 rows of your student DataFrame
df.head(2)
```

```python
# Task 5: Rename one column of your DataFrame using the `rename()` method
df.rename(columns={'Marks': 'Score'}, inplace=True)
df
```

```python
# Task 6: Create an array with values from 0 to 9 and extract all odd numbers
arr_odd = np.arange(10)[np.arange(10) % 2 != 0]
arr_odd
```

```python
# Task 7: Extract the first two rows and first two columns from a 4x4 NumPy array
arr_4x4 = np.arange(16).reshape(4, 4)
arr_4x4[0:2, 0:2]
```

```python
# Task 8: Given a DataFrame `df`, filter it to show only rows where a column 'Age' is greater than 25
df_age = pd.DataFrame({'Name': ['Shohag', 'Mehedi', 'Hridoy'], 'Age': [24, 30, 28]})
df_age[df_age['Age'] > 25]
```

```python
# Task 9: Select only the 'Name' and 'Salary' columns from an existing DataFrame
df_salary = pd.DataFrame({'Name': ['Partho', 'Rakib', 'Sakhawat'], 'Salary': [50000, 55000, 60000]})
df_salary[['Name', 'Salary']]
```

```python
# Task 10: Find and print the unique values in a specific column of your DataFrame
df_unique = pd.DataFrame({'Name': ['Partho', 'Rakib', 'Sakhawat']})
df_unique['Name'].unique()
```
