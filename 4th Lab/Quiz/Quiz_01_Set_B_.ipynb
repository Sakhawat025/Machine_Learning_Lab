```python
# Task 1: Create an array with values from 0 to 9 and extract all odd numbers
import numpy as np
arr_odd = np.arange(10)[np.arange(10) % 2 != 0]
arr_odd
```

```python
# Task 2: Extract the first two rows and first two columns from a 4x4 NumPy array
arr_4x4 = np.arange(16).reshape(4, 4)
arr_4x4[0:2, 0:2]
```

```python
# Task 3: Given a DataFrame `df`, filter it to show only rows where a column 'Age' is greater than 25
import pandas as pd
df_age = pd.DataFrame({'Name': ['Sakhawat', 'Rakib', 'Shahin'], 'Age': [24, 30, 28]})
df_age[df_age['Age'] > 25]
```

```python
# Task 4: Select only the 'Name' and 'Salary' columns from an existing DataFrame
df_salary = pd.DataFrame({'Name': ['Sakhawat', 'Rakib', 'Shahin'], 'Salary': [50000, 55000, 60000]})
df_salary[['Name', 'Salary']]
```

```python
# Task 5: Find and print the unique values in a specific column of your DataFrame
df_unique = pd.DataFrame({'Name': ['Sakhawat', 'Rakib', 'Shahin']})
df_unique['Name'].unique()
```
