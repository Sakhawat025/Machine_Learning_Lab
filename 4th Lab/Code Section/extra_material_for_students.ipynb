```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import ticker
```

```python
stackoverflow_monthly = pd.read_csv('/content/stackoverflow.csv', parse_dates=True, index_col='creation_date').loc[:'2021-08','pandas':'bokeh'].resample('1M').sum()

stackoverflow_daily = pd.read_csv( '/content/stackoverflow.csv', parse_dates=True, index_col='creation_date').loc[:,'pandas':'bokeh'].resample('1D').sum()
```

```python
subset = stackoverflow_daily.sum().nlargest(4)
top_libraries_monthly = stackoverflow_monthly.reindex(columns=subset.index)
top_libraries_monthly.head()
```

```python
def despine(ax):
    for spine in ['top', 'right']:
        ax.spines[spine].set_visible(False)
    return ax

def area_plot(data):
    fig, ax = plt.subplots(figsize=(12, 5))
    ax.stackplot(
        mdates.date2num(data.index),
        data.to_numpy().T,
        labels=data.columns
    )
    ax.set(
        xlabel='', ylabel='tagged questions',
        title='Stack Overflow Questions per Month'
    )
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    ax.yaxis.set_major_formatter(ticker.EngFormatter())
    despine(ax)
    return ax

def annotate(ax, data):
    total = 0
    for area in ax.collections:
        library = area.get_label()
        last = data.last('1D')[library]
        last_day, last_value = last.index[0], last.iat[0]
        if library != 'seaborn':
            kwargs = {}
        else:
            kwargs = dict(
                xytext=(last_day + pd.Timedelta(days=20), (last_value + total) * 1.1),
                arrowprops=dict(arrowstyle='->')
            )

        ax.annotate(
            f' {library}: {data.loc["2021", library].median():,.0f}',
            xy=(last_day, last_value / 2 + total), ha='left', va='center', **kwargs
        )
        total += last_value
    return ax
```

```python
ax = area_plot(top_libraries_monthly)
annotate(ax, top_libraries_monthly)
```

```python
def stacked_bars(data):
  fig, ax = plt.subplots(figsize=(7,4))
  libraries = data.index

  last = 0
  for library in libraries:
    co_occuring_library = data[library]
    ax.barh(libraries, co_occuring_library, label=library, left=last)
    last+=co_occuring_library

  ax.yaxis.set_inverted(True)
  ax.legend(bbox_to_anchor=(1.35, 0.5), loc='center right', framealpha=0.5)
  ax.set(xlabel='% of Qs', xlim=(0,1))
  ax.xaxis.set_major_formatter(ticker.PercentFormatter(xmax=1));
  return ax

## Annotation
def annotate_bars(ax, thres):
  for patch in ax.patches:
    width = patch.get_width()
    if width>thres:
      ax.text(
              patch.get_x()+width/2, patch.get_y()+patch.get_height()/2, f'{width:.1%} ',
              va='center', ha='center', color='white'
          )
  return ax


ax = annotate_bars(stacked_bars(co_occur), 0.09)
```

```python
def stacked_bars(data):
  fig, ax = plt.subplots(figsize=(7,4))
  libraries = data.index

  last = 0
  for i, library in enumerate(libraries):
    co_occuring_library = data[library]
    ax.barh(libraries, co_occuring_library, label=library, left=last, color=cmap(i))
    last+=co_occuring_library

  ax.yaxis.set_inverted(True)
  ax.legend(bbox_to_anchor=(1.35, 0.5), loc='center right', framealpha=0.5)
  ax.set(xlabel='% of Qs', xlim=(0,1))
  ax.xaxis.set_major_formatter(ticker.PercentFormatter(xmax=1));
  return ax
```
