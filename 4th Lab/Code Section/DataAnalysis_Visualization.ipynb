# Matplotlib

```python
import matplotlib.pyplot as plt  # Import plotting tool
%matplotlib inline

x=[1,2,3,4,5]           # Set X-axis values
y=[10,5,14,20,3]        # Set Y-axis values
plt.plot(x,y)           # Draw line
plt.xlabel('AGE')       # Label X
plt.ylabel('age')       # Label Y
plt.title('Relation')   # Add title
plt.show()              # Display
```

```python
x=[1,2,3,4,5,6,7]            # Days of the week
y=[50,51,52,45,41,47,49]     # Temperature values
plt.xlabel('Day')            # Set bottom label
plt.ylabel('Temp')           # Set side label
plt.title('Weather forecast') # Set top title

# Draw a purple, thick, dotted line
plt.plot(x,y, color='purple', linewidth=4, linestyle=':')
plt.show()                   # Display plot
```

```python
x=[1,2,3,4,5,6,7]            # List of days
y=[50,51,52,45,41,47,49]     # List of temperatures
plt.xlabel('Day')            # Label X-axis
plt.ylabel('Temp')           # Label Y-axis
plt.title('Weather forecast') # Set title
plt.plot(x,y, 'r*:')         # Plot red stars with dotted lines
plt.show()                   # Show final graph
```

```python
plt.plot(x,y, color='blue', marker='*', linestyle='') # Plot blue stars with no line
plt.show()                                            # Show only the markers

```

```python
plt.plot(x,y, color='red', marker='*', linestyle=':')
plt.show()
```

```python
days=[1,2,3,4,5,6,7]            # X-axis data (days)
max_t=[50,51,52,54,50,49,47]    # Max temp data
min_t=[40,41,42,45,39,42,41]    # Min temp data
avg_t=[45,46,47,44,43,45,48]    # Average temp data

plt.xlabel('Day')               # Label bottom axis
plt.ylabel('Temp')              # Label side axis
plt.title('weather')            # Add plot title

plt.plot(days,max_t, label='max temp', marker="*") # Plot max temp with stars
plt.plot(days,min_t, label='min temp')             # Plot min temp line
plt.plot(days,avg_t, label='avg temp')             # Plot avg temp line

plt.legend(shadow=True, fontsize='small') # Show labels with a small shadow box
plt.show()                                # Display the complete graph
```

```python
import numpy as np
company = ['google', 'facebook', 'amazon', 'tesla'] # List of names
revenue=[15,20,30,70]                               # List of values

xpos = np.arange(len(company)) # Create number positions (0,1,2,3)
plt.xticks(xpos,company)       # Replace numbers with company names
plt.bar(xpos, revenue, label='revenue') # Draw bars
plt.legend()                   # Show the label box
plt.show()                     # Display chart
```

```python
profit=[10,15,25,65]                                     # New data for profit
plt.bar(xpos-0.2, revenue, width=0.4, label='revenue')   # Shift revenue bars to the left
plt.bar(xpos+0.2, profit, width=0.4, label='profit')     # Shift profit bars to the right
plt.xticks(xpos,company)                                 # Name the companies on X-axis
plt.legend(shadow=True)                                  # Show legend with shadow
plt.show()                                               # Display side-by-side bars
```

```python
plt.barh(xpos, revenue)    # Create horizontal bars
plt.yticks(xpos, company)  # Label the Y-axis with company names
plt.show()                 # Display plot
```

```python
flights = sns.load_dataset('flights').pivot(index="month", columns='year')
flights
```

```python
plt.title("No. of passengers") # Add top title
sns.heatmap(flights)           # Create color-coded grid
plt.show()                     # Display map
```

```python
fig, ax = plt.subplots(figsize=(10, 15)) # Set a large figure size
sns.heatmap(flights, fmt=".1f", annot=True, cmap="Blues", ax=ax) # Create blue heatmap with rounded numbers
ax.set_title("No. of passengers")        # Add title to the axis
plt.show()                               # Display the final plot
```

```python
plt.figure(figsize=(8,8))                     # Set plot to a square 8x8 size
plt.rcParams.update({'font.size':10})         # Set text size for the plot
sns.heatmap(flights, fmt='d', annot=True,     # Plot whole numbers with labels
            cmap='Blues', vmin=100, vmax=622, # Set color range and blue theme
            square=True, linewidth=1.4)       # Make cells square with borders
plt.show()                                    # Display final heatmap
```

```python
plt.title("Flower")                                              # Add title
sns.kdeplot(data=flw, x='sepal_length', y='sepal_width',
            shade=True, shade_lowest=False)                      # Draw shaded density contours
sns.lmplot(data=flw, x='sepal_length', y='sepal_width',
           fit_reg=False)                                        # Draw scatter plot without trend line
plt.show()                                                       # Display result
```

```python
plt.title("Flower")                                              # Set chart title
# Plot Setosa density in Red
sns.kdeplot(data = flw[flw.species=='setosa'], x='sepal_length', y='sepal_width', shade=True, shade_lowest=False, cmap='Reds')
# Plot Virginica density in Blue
sns.kdeplot(data = flw[flw.species=='virginica'], x='sepal_length', y='sepal_width', shade=True, shade_lowest=False, cmap='Blues')
# Plot Versicolor density in Green
sns.kdeplot(data = flw[flw.species=='versicolor'], x='sepal_length', y='sepal_width', shade=True, shade_lowest=False, cmap='Greens')
plt.show()                                                       # Display overlapping density map
```
