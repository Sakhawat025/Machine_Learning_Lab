When using drop_first=True, the first category (alphabetically) is treated as baseline to avoid redundancy. Any missing column (like salary_high) is inferred when all other dummy columns are 0.

**Why Drop the First Column?**

Avoid dummy variable trap in linear models. If you keep all 3 columns, they are linearly dependent:

salary_low + salary_medium + salary_high = 1

This can cause multicollinearity in regression.
